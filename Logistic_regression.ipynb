# Output variable -> y
y -> Whether the client has subscribed a term deposit or not 
Binomial ("yes" or "no")

```python
#import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score as ac
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

```python
#importing dataset
bankdata=pd.read_csv('bank-full.csv',sep=';')
bankdata.tail(20)
```

# EDA

```python
bankdata.shape
```

```python
bankdata.info()
```

```python
bankdata.head(2)
```

```python
#customizing Hot encoding to the  data
data1=pd.get_dummies(bankdata,columns=['job','marital','education','contact','month','poutcome'])
data1.head()
```

```python
data1.info()
```

```python
#Customizing binary encoding to data
data1['default'] = np.where(data1['default'].str.contains("yes"), 1, 0)
data1['housing'] = np.where(data1['housing'].str.contains("yes"), 1, 0)
data1['loan'] = np.where(data1['loan'].str.contains("yes"), 1, 0)
data1['y'] = np.where(data1['y'].str.contains("yes"), 1, 0)
data1
```

# Building Model

```python
#splitting input variables and output variables
X=pd.concat([data1.iloc[:,0:10],data1.iloc[:,11:-1]],axis=1)
X.info()
```

```python
Y=data1.iloc[:,10]
Y.info()
```

```python
#Splitting Data into train test
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.25,shuffle=True)
```

```python
classifier = LogisticRegression()
```

```python
classifier=classifier.fit(xtrain,ytrain)
```

# Model Prediction

```python
#predicting xtest data value to the model
x_predict=classifier.predict(xtest)
x_predict
```

# Testing Model

```python
#Checking accuracy of ytest actual vs ytest prediction
ac(ytest,x_predict)
```

```python
#confusion matrix for the model accuracy
confusion_matrix(ytest,x_predict)
```

```python
#Classification report
print(classification_report(ytest,x_predict))
```
